# Data Cleaning

Removing Unnamed column.

```python
df_b.drop(['Unnamed: 0', 'round', 'day', 'time'], axis = 1, inplace = True)
df_b.head()
```

Check for NaN.

```python
df_b.isna().sum()
```

We will keep bundesliga data as it is. 

This function fills the attendance of a stadium with the average attendance only when the value is NaN for matches outside the Covid period. 

```python
def att_clean(s, att_map):
    for index, value in s.attendance.iteritems():
        if np.isnan(value) and not (datetime.datetime(2020, 4, 1) < s.date[index] < datetime.datetime(2021, 8, 1)):
            s.at[index, 'attendance'] = att_map.get(s.at[index, 'venue'])
```

```python
l_att_map = df_l.groupby('venue').mean()['attendance'].apply(lambda x: x if np.isnan(x) else int(math.floor(x)) ).to_dict()

#that season rm had 0 attendance 
l_att_map['Estadio Alfredo Di Stéfano'] = 0
l_att_map
```

```python
att_clean(df_l, l_att_map)
df_l.isna().sum()
```

Cleaning attendance for the Premier League.

```python
p_att_map = df_p.groupby('venue').mean()['attendance'].apply(lambda x: x if np.isnan(x) else int(math.floor(x)) ).to_dict()
p_att_map
```

Cleaning attendance for the Serie A.

```python
s_att_map = df_s.groupby('venue').mean()['attendance'].apply(lambda x: x if np.isnan(x) else int(math.floor(x)) ).to_dict()
s_att_map
```

```python
att_clean(df_s, s_att_map)
df_s.isna().sum()
```

Adding a goals column.

```python
g_home = []
g_away = []
```

```python
def score(s):
    g_home.append(int(s[0]))
    g_away.append(int(s[2]))
    return s
def score_clean(df):
    df.score.apply(score)
```

```python
score_clean(df_b)
df_b['g_home'] = g_home
df_b['g_away'] = g_away
df_b.head()
```

```python
score_clean(df_l)
df_l['g_home'] = g_home
df_l['g_away'] = g_away
df_l.head()
```

```python
score_clean(df_p)
df_p['g_home'] = g_home
df_p['g_away'] = g_away
df_p.head()
```

```python
score_clean(df_s)
df_s['g_home'] = g_home
df_s['g_away'] = g_away
df_s.head()
```

Adding a results column.

```python
def result(df):
    result = []
    for index in range(len(df)):
        if df['g_home'][index] > df['g_away'][index]:
            result.append(df['home'][index])
        elif df['g_home'][index] < df['g_away'][index]:
            result.append(df['away'][index])
        else:
            result.append('Draw')
    return result
```

```python
df_b['result'] = result(df_b)
df_b.head()
```

```python
df_l['result'] = result(df_l)
df_l.head()
```

```python
df_p['result'] = result(df_p)
df_p.head()
```

```python
df_s['result'] = result(df_s)
df_s.head()
```
